# Data Visualization

## FacetGrid

Small Multiples: 

```python
import seaborn as sns
from matplotlib  import pyplot as plt
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
sns.set_style('white')
```

```python
g = sns.FacetGrid(penguins, col='island')

# map some plots into the places
# You can use .map or .map_dataframe
g.map_dataframe(sns.histplot, x="flipper_length_mm")

```

```python
type(g)
```

```python
# methods & attributes
# sharey=False will give each chart y and x values
# ylim=(x,y) will fix y values

g = sns.FacetGrid(penguins, col='island', row='species', sharey=False)

# map some plots into the places
# scatterplot requires at least 2 arguments
g.map_dataframe(sns.scatterplot, x="bill_depth_mm", y="bill_length_mm")
g.set_axis_labels("label1", "label2")

g.set_titles(col_template="{col_name} Island")
```
