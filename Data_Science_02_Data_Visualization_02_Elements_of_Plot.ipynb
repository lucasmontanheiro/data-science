#Practical Data Visualization

##Segment 2 - Defining Elements of a Plot

Plot elements add context to your plot, so the plot effectively communicates

There's two ways:
1. Functional method
2. Object-oriented method 

Steps in Object-Oriented Plotting
1. Create a blank figure object
2. Add axes to the figure
3. Generate plot(s) within figure object
4. Specify plotting and layout parameters for the plots within your figure

Subplot
- Is a plotting figure that contains more than one plot, or subplots.

```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams["figure.figsize"] = 5,4
```

**Defining axes, ticks and grids**

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure() #Creating a blank figure
ax = fig.add_axes([.1,.1,1,1]) #what we plug along the axes

ax.plot(x,y)
```

```python
fig = plt.figure() #Creating a blank figure
ax = fig.add_axes([.1,.1,1,1]) #what we plug along the axes

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

ax.plot(x,y)
```

```python
##Adding a grid to the plot

fig = plt.figure() #Creating a blank figure
ax = fig.add_axes([.1,.1,1,1]) #what we plug along the axes

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()
ax.plot(x,y)
```

**Generating multiple plots in one figure with subplots**

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.plot(x)
ax2.plot(x,y)
```
