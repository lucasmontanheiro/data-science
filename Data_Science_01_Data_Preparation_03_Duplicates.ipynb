#Remove duplicates




```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

**Removing duplicates**

```python
DF_obj = DataFrame({'column 1':[1, 1, 2, 2, 3, 3, 3],
                    'column 2':['a', 'a', 'b', 'b', 'c', 'c', 'c'],
                    'column 3':['A', 'A', 'B', 'B', 'C', 'C', 'C']})
DF_obj
```

```python
DF_obj.duplicated()
```

```python
#Drop duplicate ROWS
DF_obj.drop_duplicates()
```

```python
#Drop duplicate COLUMNS
DF_obj_b = DataFrame({'column 1':[1, 1, 2, 2, 3, 3, 3],
                    'column 2':['a', 'a', 'b', 'b', 'c', 'c', 'c'],
                    'column 3':['A', 'A', 'B', 'B', 'C', 'D', 'C']})
DF_obj_b.drop_duplicates(['column 3'])
```
