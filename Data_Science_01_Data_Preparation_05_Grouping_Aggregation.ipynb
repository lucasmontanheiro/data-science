#Data Preparation

###Grouping & Aggregation

```python
import pandas as pd
import numpy as np

from pandas import Series, DataFrame
```

```python
address = 'https://raw.githubusercontent.com/lucasmontanheiro/db/main/exercises/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ["cars_names", "mpg", "cyl", "disp", "hp", "drat", "wt", "qsec", "vs", "am", "gear", "carb"]
cars.head()
```

```python
# Read median value for a specific column
cars["wt"].median()

# Read median value for all numeric columns
cars.median()
```

**COUNT vs COUNT VALUES functions**

```python
# How many cars there are in the dataset
cars["gear"].count()
```

```python
# How many cars have 3, 4 or 5 gears
cars["gear"].value_counts()
```

**GROUP BY**

Pandas documentation says group by operations involves:
1. Split
2. Apply function
3. Combine results

```python
gear_group = cars.groupby("gear")
```

```python
gear_group.get_group(4)
```

```python
filt = cars["gear"] == 5
cars.loc[filt]
```

```python
# Count values by 2 filters: GEAR and AM
gear_group["am"].value_counts()
```
