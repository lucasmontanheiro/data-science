#Treating missing values
*  By default, missing values are represented with NaN: "Not a Number"
*  Warning: if you dataset has 0s, 99s etc you need to drop and clean the dataset

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

**Figuring out what data is missing**

```python
missing = np.nan
series_obj = Series(['row 1', 'row 2', missing, 'row 4', 'row 5', 'row 6', missing, 'row 8'])
series_obj  
```

```python
series_obj.isnull()
```

**Filling in for missing values**

```python
np.random.seed(25)
DF_obj = DataFrame(np.random.rand(36).reshape(6,6))
DF_obj
```

```python
DF_obj.loc[3:5, 0] = missing
DF_obj.loc[1:4, 5] = missing
DF_obj
```

```python
#Counting missing values
DF_obj.isnull().sum()
```

```python
filled_DF = DF_obj.fillna(0)
filled_DF
```

```python
filled_DF = DF_obj.fillna({0: 0.1, 5:1.25})
filled_DF
```

```python
# method='ffill' fills the number with the previous data in the column
fill_DF = DF_obj.fillna(method='ffill')
fill_DF
```

Filtering out missing values

```python
#Dropping the ROWS with NA values
DF_no_NaN = DF_obj.dropna()
DF_no_NaN
```

```python
#Dropping the COLUMNS with NA values
DF_no_NaN_c = DF_obj.dropna(axis=1)
DF_no_NaN_c
```
