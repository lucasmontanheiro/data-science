#Data Preparation

###Melt

```python
import pandas as pd
import seaborn as sns
```

```python
weather = pd.read_csv("https://raw.githubusercontent.com/lucasmontanheiro/db/83fd249e1a90e2012f5b37bc6c5785a868ded49b/exercises/weather.csv")
```

```python
weather
```

```python
melted_data = pd.melt(weather, id_vars=["day"])
melted_data
```

```python
g = sns.catplot(x="day", y="value", hue='variable', data=melted_data, kind='point')
```
